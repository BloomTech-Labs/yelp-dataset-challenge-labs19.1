```python
# stars is an integer that is the star rating that we want statistics for.
# subjectivity is a list of the ranges of subjectivity to gather statistics for.
# subjectivity must have an even length, and no number can be below 0, or greater than 1.

def sentimentstats(stars = 0, subjectivity = [0,1]):
    print('These are the statistics for ' + str(stars) + ' stars, and Subjectivity between ' + str(subjectivity[0]) + ' and ' + str(subjectivity[1]) + '.')
    plt.plot(sandrdf.loc[sandrdf['stars'] == stars].loc[sandrdf['subjectivity'] > subjectivity[0]].loc[sandrdf['subjectivity'] < subjectivity[1], 'polarity'].reset_index().drop('index',1))
    plt.show()
    print(sandrdf.loc[sandrdf['stars'] == stars].loc[sandrdf['subjectivity'] > subjectivity[0]].loc[sandrdf['subjectivity'] < subjectivity[1], 'polarity'].reset_index().drop('index',1).describe())
```

```python
subjec = [[0,0.1], [0.1,0.2], [0.2,0.3], [0.3,0.4],[0.4,0.5],[0.5,0.6],[0.6,0.7],[0.7,0.8],[0.8,0.9],[0.9,1]]
for i in range(1,6):
    for x in subjec:
        sentimentstats(i,x)
```
